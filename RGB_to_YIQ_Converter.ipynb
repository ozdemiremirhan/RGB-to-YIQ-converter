```python
import numpy as np
import matplotlib.pylab as plt
import cv2
import matplotlib.image as mpimg
```

```python
# Read Image
img1=cv2.imread('color_template.png')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
# Resize image
img1=cv2.resize(img1,(224,224))
plt.imshow(img1)
```

```python
# Define the RGB to YUV transformation matrix
RGB2YUV_matrix=np.array([[0.2126, 0.7152, 0.0722],
                           [-0.09991, -0.33609, 0.436],
                           [0.615, -0.55861, -0.05639]])
img1_row=len(img1)
img1_col=len(img1[0])

# Convert image from RGB to YIQ
img1_yuv = np.zeros_like(img1, dtype=float)
for i in range(img1_row):
  for j in range(img1_col):
    R,G,B=img1[i,j]/255  # Normalize RGB values
    RGB_vector=np.array([R,G,B])
    new_pixel=np.dot(RGB2YUV_matrix,RGB_vector)
    img1_yuv[i, j] = new_pixel

# Display the YUV image
plt.imshow(img1_yuv)
```

```python

# Define the YUV to RGB transformation matrix
YUV2RGB_matrix = np.array([[1, 0, 1.28033],
                           [1, -0.21482, -0.38059],
                           [1, 2.12798, 0]])

# Create an empty array for RGB images
img1_rgb = np.zeros_like(img1)

# Convert each pixel from YUV to RGB
for k in range(224):
    for l in range(224):
        Y, U, V = img1_yuv[k, l]
        YUV_vector = np.array([Y, U, V])
        new_pixel2 = np.dot(YUV2RGB_matrix, YUV_vector)
        img1_rgb[k, l] = np.clip(new_pixel2 * 255, 0, 255)  # Clip to ensure values are within valid range

# Display the converted RGB image
plt.imshow(img1_rgb)
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the cat and dog images
cat_image1 = cv2.imread('cat1.jpg')
cat_image2 = cv2.imread('cat2.jpg')
cat_image3 = cv2.imread('cat3.jpg')
dog_image1 = cv2.imread('dog1.jpg')

# Convert color BGR to RGB
cat_image1=cv2.cvtColor(cat_image1,cv2.COLOR_BGR2RGB)
cat_image2=cv2.cvtColor(cat_image2,cv2.COLOR_BGR2RGB)
cat_image3=cv2.cvtColor(cat_image3,cv2.COLOR_BGR2RGB)
dog_image1=cv2.cvtColor(dog_image1,cv2.COLOR_BGR2RGB)

# Flipping vertically
cat1_flipped_vertical = cv2.flip(cat_image1, 0)
cat2_flipped_vertical = cv2.flip(cat_image2, 0)
cat3_flipped_vertical = cv2.flip(cat_image3, 0)
dog1_flipped_vertical = cv2.flip(dog_image1, 0)

# Flipping horizontally
cat1_flipped_horizontal = cv2.flip(cat_image1, 1)
cat2_flipped_horizontal = cv2.flip(cat_image2, 1)
cat3_flipped_horizontal = cv2.flip(cat_image3, 1)
dog1_flipped_horizontal = cv2.flip(dog_image1, 1)

# Rotate left by 90 degrees
cat1_rotated_left = cv2.rotate(cat_image1, cv2.ROTATE_90_COUNTERCLOCKWISE)
cat2_rotated_left = cv2.rotate(cat_image2, cv2.ROTATE_90_COUNTERCLOCKWISE)
cat3_rotated_left = cv2.rotate(cat_image3, cv2.ROTATE_90_COUNTERCLOCKWISE)
dog1_rotated_left = cv2.rotate(dog_image1, cv2.ROTATE_90_COUNTERCLOCKWISE)

# Rotate right by 90 degrees
cat1_rotated_right = cv2.rotate(cat_image1, cv2.ROTATE_90_CLOCKWISE)
cat2_rotated_right = cv2.rotate(cat_image2, cv2.ROTATE_90_CLOCKWISE)
cat3_rotated_right = cv2.rotate(cat_image3, cv2.ROTATE_90_CLOCKWISE)
dog1_rotated_right = cv2.rotate(dog_image1, cv2.ROTATE_90_CLOCKWISE)

# Resize to half
cat1_resized = cv2.resize(cat_image1, (int(cat_image1.shape[1] / 2), int(cat_image1.shape[0] / 2)))
cat2_resized = cv2.resize(cat_image2, (int(cat_image2.shape[1] / 2), int(cat_image2.shape[0] / 2)))
cat3_resized = cv2.resize(cat_image3, (int(cat_image3.shape[1] / 2), int(cat_image3.shape[0] / 2)))
dog1_resized = cv2.resize(dog_image1, (int(dog_image1.shape[1] / 2), int(dog_image1.shape[0] / 2)))

# Display images
plt.imshow(cat_image1)
plt.title('Cat1 Original')
plt.show()

plt.imshow(cat1_flipped_vertical)
plt.title('Cat1 Flipped Vertical')
plt.show()

plt.imshow(cat1_flipped_horizontal)
plt.title('Cat1 Flipped Horizontal')
plt.show()

plt.imshow(cat1_rotated_left)
plt.title('Cat1 Rotated Left')
plt.show()

plt.imshow(cat1_rotated_right)
plt.title('Cat1 Rotated Right')
plt.show()

plt.imshow(cat1_resized)
plt.title('Cat1 Resized')
plt.show()

plt.imshow(cat_image2)
plt.title('Cat2 Original')
plt.show()

plt.imshow(cat2_flipped_vertical)
plt.title('Cat2 Flipped Vertical')
plt.show()

plt.imshow(cat2_flipped_horizontal)
plt.title('Cat2 Flipped Horizontal')
plt.show()

plt.imshow(cat2_rotated_left)
plt.title('Cat2 Rotated Left')
plt.show()

plt.imshow(cat2_rotated_right)
plt.title('Cat2 Rotated Right')
plt.show()

plt.imshow(cat2_resized)
plt.title('Cat2 Resized')
plt.show()

plt.imshow(cat_image3)
plt.title('Cat3 Original')
plt.show()

plt.imshow(cat3_flipped_vertical)
plt.title('Cat3 Flipped Vertical')
plt.show()

plt.imshow(cat3_flipped_horizontal)
plt.title('Cat3 Flipped Horizontal')
plt.show()

plt.imshow(cat3_rotated_left)
plt.title('Cat3 Rotated Left')
plt.show()

plt.imshow(cat3_rotated_right)
plt.title('Cat3 Rotated Right')
plt.show()

plt.imshow(cat3_resized)
plt.title('Cat3 Resized')
plt.show()

plt.imshow(dog_image1)
plt.title('Dog1 Original')
plt.show()

plt.imshow(dog1_flipped_vertical)
plt.title('Dog1 Flipped Vertical')
plt.show()

plt.imshow(dog1_flipped_horizontal)
plt.title('Dog1 Flipped Horizontal')
plt.show()

plt.imshow(dog1_rotated_left)
plt.title('Dog1 Rotated Left')
plt.show()

plt.imshow(dog1_rotated_right)
plt.title('Dog1 Rotated Right')
plt.show()

plt.imshow(dog1_resized)
plt.title('Dog1 Resized')
plt.show()
```
